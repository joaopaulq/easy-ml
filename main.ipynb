```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets, linear_model
from sklearn.preprocessing import StandardScaler

from util import add_intercept
from linear_reg import LinearRegression
from logistic_reg import LogisticRegression
from perceptron import Perceptron
from gda import GaussianDiscriminantAnalysis

%matplotlib inline
```

# Regressão Linear
### Pré-processamento

```python
# Carrega o conjunto de dados.
x, y = datasets.load_diabetes(return_X_y=True)
# Normaliza todo conjunto de dados.
x = StandardScaler().fit_transform(x)
# Usa apenas 1 feature.
x = x[:, np.newaxis, 2]
# Adiciona o termo de interseção a matriz x.
x_inter = add_intercept(x)
```

### Implementado

```python
lin = LinearRegression()
lin.fit(x_inter, y)
print(f"coef: {lin.theta[1:]} \nintercept: {lin.theta[0]}")
```

### scikit-learn

```python
sk_lin = linear_model.LinearRegression()
sk_lin.fit(x, y)
print(f"coef: {sk_lin.coef_} \nintercept: {sk_lin.intercept_}")
```

### Plot

```python
plt.figure(figsize=(12,7))
plt.scatter(x, y, color='black')
plt.plot(x, lin.predict(x_inter), color='blue', ls='-', linewidth=4, label='my')
plt.plot(x, sk_lin.predict(x), color='red', ls=':', linewidth=4, label='sklearn')
plt.legend()
plt.show()
```

# Perceptron
### Pré-processamento

```python
# Carrega o conjunto de dados.
x, y = datasets.load_breast_cancer(return_X_y=True)
# Normaliza todo conjunto de dados.
x = StandardScaler().fit_transform(x)
# Usa apenas 1 feature.
x = x[:, np.newaxis, 3]
# Adiciona o termo de interseção a matriz x.
x_inter = add_intercept(x)
```

### Implementado

```python
per = Perceptron(lr=0.01, max_iter=10000, verbose=False)
per.fit(x_inter, y)
print(f"coef: {per.theta[1:]} \nintercept: {per.theta[0]}")
```

### scikit-learn

```python
sk_per = linear_model.Perceptron()
sk_per.fit(x, y)
print(f"coef: {sk_per.coef_} \nintercept: {sk_per.intercept_}")
```
