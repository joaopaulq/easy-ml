# Example - Random Blobs

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=500,
                  centers=2,
                  cluster_std=0.60,
                  random_state=0)

plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis')
```

```python
knn = KNN()
knn.fit(X, y)
```

```python
h_x = knn.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=h_x, s=50, cmap='viridis')
```
