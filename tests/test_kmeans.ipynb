# Setup

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from models.kmeans import KMeans
```

# Example 1 - Random Data

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=300,
                  centers=4,
                  cluster_std=0.60,
                  random_state=0)

plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis')
```

```python
km = KMeans(k=4)
km.fit(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=km.clusters, s=50, cmap='viridis')
plt.scatter(km.centroids[:, 0], km.centroids[:, 1], marker='x', c='red')
```

# Example 2 - MNIST

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
km = KMeans(k=10)
km.fit(digits.data)
```

```python
fig, ax = plt.subplots(2, 5)
ax = ax.flatten()
for i in range(10):
    plottable_image = np.reshape(km.centroids[i], (8, 8))
    ax[i].imshow(plottable_image, cmap='gray_r')
```
