# Example - Random Blobs

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=300,
                  centers=2,
                  cluster_std=0.60,
                  random_state=0)

plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis')
```

## Perceptron

```python
per = Perceptron()
per.fit(X, y)
```

```python
h_x = per.predict(X)
J = per.loss(h_x, y)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=h_x, s=50, cmap='viridis')
print(f"Loss {J}")
```
