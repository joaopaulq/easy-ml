```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from gda import GDA
```

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=500,
                  centers=2,
                  cluster_std=0.60,
                  random_state=0)

plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis')
```

```python
g = GDA()
g.fit(X, y)
```

```python
g.sigma.shape
```

```python
h_x = g.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=h_x, s=50, cmap='viridis')
```
